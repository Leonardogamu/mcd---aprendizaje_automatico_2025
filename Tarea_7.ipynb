#Tarea 7

```python
import pandas as pd

data=pd.read_csv('/content/health_lifestyle_dataset.csv')
data['gender_bin'] = data['gender'].map({'Male': 0, 'Female': 1})
data['gender'] = data['gender_bin']
data.drop('gender_bin', axis=1, inplace=True) #recordamos la modificacion hecha en tareas previas

data
```

```python
from sklearn.preprocessing import MinMaxScaler
data_x = data[[
    #'id',
    'age',
    'gender',
    'bmi',
    'daily_steps',
    'sleep_hours',
    'water_intake_l',
    'calories_consumed',
    'smoker',
    'alcohol',
    'resting_hr',
    'systolic_bp',
    'diastolic_bp',
    'cholesterol',
    'family_history',
    'disease_risk'
]]
df_t = data_x.dropna().head(2000)

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_t)
df_scaled = pd.DataFrame(scaled, columns=['age',
    'gender',
    'bmi',
    'daily_steps',
    'sleep_hours',
    'water_intake_l',
    'calories_consumed',
    'smoker',
    'alcohol',
    'resting_hr',
    'systolic_bp',
    'diastolic_bp',
    'cholesterol',
    'family_history',
    'disease_risk'])
df_scaled
```

```python
#Separamos en entrenamiento y prueba
df_train = df_scaled.sample(frac = 0.7)
df_train
```

```python
df_test = df_scaled.drop(df_train.index)
df_test
```

```python
X_train = df_train[['age',
    'gender',
    'bmi',
    'daily_steps',
    'sleep_hours',
    'water_intake_l',
    'calories_consumed',
    'smoker',
    'alcohol',
    'resting_hr',
    'systolic_bp',
    'cholesterol',
    'family_history',
    'disease_risk']]
y_train = df_train[['diastolic_bp']]

X_test = df_test[['age',
    'gender',
    'bmi',
    'daily_steps',
    'sleep_hours',
    'water_intake_l',
    'calories_consumed',
    'smoker',
    'alcohol',
    'resting_hr',
    'systolic_bp',
    'cholesterol',
    'family_history',
    'disease_risk']]
y_test = df_test[['diastolic_bp']]
```

```python
y_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
```

```python
tree_reg.fit(X_train, y_train)
```

```python
y_pred = tree_reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
df_scaled_def = df_scaled.drop('diastolic_bp', axis=1)
df_scaled_def
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(28, 16))
plot_tree(tree_reg, feature_names=df_scaled_def.columns,
          filled=True, rounded=True,
          proportion=True,
          impurity=False,
          fontsize=10)
plt.title("Árbol de Decisión (Bajo Regresión)", fontsize=20, fontweight='bold', color="#2C3E50")
plt.tight_layout(pad=3.0)
plt.show()
```

##Okey, entendiendo el algoritmo de los arboles de decisiones, podemos proceder a un Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

rf_reg = RandomForestRegressor(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

```

```python
rf_reg.fit(X_train, y_train)
```

```python
y_pred_rf = rf_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,y_pred_rf)
mae
```

```python
# Graficar uno de los árboles del bosque
one_tree = rf_reg.estimators_[0]
plt.figure(figsize=(24, 12))
plot_tree(one_tree, feature_names=df_scaled_def.columns, filled=True, rounded=True, fontsize=10)
plt.title("Un Árbol dentro del Random Forest (Regresión con MAE)", fontsize=16)
plt.show()
```
