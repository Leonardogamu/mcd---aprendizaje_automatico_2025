## ○ Crear, modificar o eliminar columnas

```python
#No veo necesario crear columnas nuevas debido a que la informacion necesaria ya se encuentra ahi
#Pero si veo necesario modificar la columna de Genero (gender)
#Esto para que tome valores binarios y no sea la unica columna del Dataframe con texto

data['gender_bin'] = data['gender'].map({'Male': 0, 'Female': 1})
#como me daba error al hacer la modificacion sobre gender, la hice en una columna nueva y luego la reemplaze con la nueva
data['gender'] = data['gender_bin']
data.drop('gender_bin', axis=1, inplace=True) #eliminamos la columna para no tener duplicadas
#ya se ve el cambio al hacer el display del dataframe
data
```

```python

#Adicionalmente voy a verificar si no hay nulls o na, para ver si se les puede dar un tratamiento a aquellas columnas
#que los presenten
data.isnull().sum()
```

```python
data.isna().sum()
```

Como observamos no hay 'null' ni 'na' en nuestras columnas, lo cual es un buen indicio de que no habra que limpiar las columnas por ese lado.

## ○ Filtrar registros

```python
#Vamos a filtrar varias columnas para ver como se reflejan en el dataframe
data.query('smoker == 1 & alcohol == 1')
#obtenemos 6074 registros de personas que si fuman y si consumen alcohol
#teniendo un 6% de personas que consumen ambas sustancias en total de todo nuestro dataset
```

```python
data.query('daily_steps > 10000')
#Ahora obtenemos 52440 personas que realizan mas de 10 mil pasos diarios
#numero el cual diversos estudios indican, son los pasos necesarios para considerarse una persona activa
```

--------------------------------------------------------------------------

## ○ Crear subconjutos de datos

```python
#Vamos a crear 3 subconjuntos de datos
# 1. Aquellos Masculinos que duerman mas de 8 horas
data_male_sleep8 = data.query('gender == 0 & sleep_hours >= 8')
data_male_sleep8
```

```python
# 2. Aquellas personas con riesgo de enfermedad relacionado a su estilo de vida
data_with_disease = data.query('disease_risk == 1')
data_with_disease
```

```python
# 3. Aquellos Masculinos que tienen riesgo de enfermedad relacionado al estilo de vida
data_male_with_disease = data.query('gender == 0 & disease_risk == 1')
data_male_with_disease
```

## ○ Crear grupos de datos



```python
#Vamos a crear 2 Grupos de datos
#1. Vamos a agrupar por genero, el numero de personas que consumen alcohol
data.groupby('gender')['alcohol'].sum()
#obtenemos que los hombres, en este dataset son quienes mas consumen alcohol
```

```python
#2. Vamos a agrupar por genero, habito de fumar y tambien de consumir alcohol, el promedio
# de la presion sistólica (en personas jovenes debe ser de 120 para considerarse presion 'sana')
data.groupby(['gender', 'smoker','alcohol'])['systolic_bp'].mean()
#el resultado no es tan alentador, vemos que no hay mucha diferencia en la media si se compara
#a una persona que consume ambas sustancias, con aquella que no.
```
